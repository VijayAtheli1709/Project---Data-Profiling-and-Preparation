# Data  Cleansing

### Dropping Class Column from the df_Business dataset.

```python
df_Business.drop(['Class'], axis=1, inplace=True)
df_Business = df_Business.reset_index(drop=True)  # To reset the df_Business index
df_Business
```

```python
df_Business.columns
```

### Replacing values

```python
df_Business.isnull().sum()
```

```python
df_Business['gender'] = df_Business['gender'].replace(['M', 'F', 'Fe male'],['Male', 'Female', 'Female'])
df_Business['customer_type'] = df_Business['customer_type'].replace(['Loyal','disloyal'],['Loyal Customer','Disloyal Customer'])
df_Business['type_of_travel'] = df_Business['type_of_travel'].replace(['Business','Business travel'],['Business Travel','Business Travel'])
df_Business['satisfaction'] = df_Business['satisfaction'].replace(['dissatisfied', 'neutral', 'satisfied'], ['Dissatisfied', 'Neutral', 'Satisfied'])
```

## Handling missing values in all columns

#### Number of missing values in each column after dropping fully duplicate observations

```python
df_Business.isnull().sum()
```

#### Droping the least number of missing values data

```python
columns_to_drop_na = ['gender', 'customer_type', 'type_of_travel', 
                       'inflight_wifi_service', 'departure/arrival_time_convenient']

df_Business.dropna(subset=columns_to_drop_na, inplace=True)
```

```python
df_Business.isnull().sum()
```

#### Handling the missing values with Mode

```python
cols = df_Business.columns
for col in cols:
    print(col)
    value_counts = df_Business[col].value_counts()
    print(value_counts, '\n\n=======================================================')
```

```python
columns_to_impute_mode = ['ease_of_online_booking', 'food_and_drink', 'inflight_entertainment', 'baggage_handling', 'cleanliness', 'satisfaction']

for col in columns_to_impute_mode:
    mode_value = df_Business[col].mode()[0]
    df_Business.fillna({col: mode_value}, inplace=True)
```

```python
df_Business.isnull().sum()
```

#### Handling the missing values with Mean

```python
df_Business['age'] = pd.to_numeric(df_Business['age'], errors='coerce')
df_Business['flight_distance'] = pd.to_numeric(df_Business['flight_distance'], errors='coerce')
df_Business['departure_delay_in_minutes'] = pd.to_numeric(df_Business['departure_delay_in_minutes'], errors='coerce')
df_Business['arrival_delay_in_minutes'] = pd.to_numeric(df_Business['arrival_delay_in_minutes'], errors='coerce')

# Setting conditions for outlier removal
# Remove rows where age is not between 0 and 100
df_Business = df_Business[(df_Business['age'] > 0) & (df_Business['age'] < 100)]

# Assuming a the longest flight path ever acheived as upper limit for flight distance;
df_Business = df_Business[(df_Business['flight_distance'] <= 9500) & (df_Business['flight_distance'] >= 1)]

# Remove rows where departure delay and arrival delay are negative
df_Business = df_Business[(df_Business['departure_delay_in_minutes'] >= 0) & (df_Business['arrival_delay_in_minutes'] >= 0)]
columns_to_impute_mean = ['age', 'flight_distance', 'departure_delay_in_minutes', 'arrival_delay_in_minutes']

for col in columns_to_impute_mean:
    mean_value = df_Business[col].mean()
    df_Business.fillna({col: mean_value}, inplace=True)
```

```python
df_Business.isnull().sum()
```

```python
cols = df_Business.columns
for col in cols:
    print(col)
    value_counts = df_Business[col].value_counts()
    print(value_counts, '\n\n=======================================================')
```

```python
columns_to_int_Datatype = ['inflight_wifi_service', 
    'departure/arrival_time_convenient', 
    'ease_of_online_booking', 
    'food_and_drink', 
    'inflight_entertainment', 
    'cleanliness','age','baggage_handling']

for col in columns_to_int_Datatype:
    df_Business[col] = df_Business[col].astype(int)
```

```python
# Columns with rating scales where 0 is rated can be replaced with lowest rating value i.e., 1
columns_to_correct = [
    'inflight_wifi_service', 
    'departure/arrival_time_convenient', 
    'ease_of_online_booking', 
    'food_and_drink', 
    'inflight_entertainment', 
    'cleanliness'
]
for col in columns_to_correct:
    df_Business[col] = df_Business[col].replace(0, 1)
```

```python
# Columns with rating scales where 10 is rated can be replaced with highest rating value i.e., 5
columns_to_correct = ['inflight_entertainment']
for col in columns_to_correct:
    df_Business[col] = df_Business[col].replace(10, 5)
```

```python
df_Business = df_Business[df_Business['satisfaction'].isin(['Satisfied', 'Dissatisfied', 'Neutral'])]
```
